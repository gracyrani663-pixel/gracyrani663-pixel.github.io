```python
import sympy as sp
X=sp.Symbol('X')
f=((x**2-1)/(x-1))
a=1
LHL=sp.limit(f,x,a,dir='-')
RHR=sp.limit(f,x,a,dir='+')
print("checking continuity at x=",a)
print ("left-hand limit =",LHL)
print("right-hand limit=",RHR)
if LHL==RHR:
   print(f"function of continous at x={a}")
else:
   print(f"function is not continous at x={a}")
```

```python
import sympy as sp
x=sp.Symbol ('x')
f=x**3-2*x-5
df=sp.diff(f,x)
f=sp.lambdify(x,f)
df=sp.lambdify(x,df)
xo=2.5
for i in range (5):
  xo=xo-f(xo)/df(xo)
  print(f"iteration {i}: x = {xo}")
```
